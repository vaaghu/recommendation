```python
import pandas as pd

chunk_size = 20000  # Adjust the chunk size as needed
chunks = pd.read_csv('anime.csv', chunksize=chunk_size)
# length = sum([1 for _ in chunks])
# print(f"Length: {length}\nitems: < {length * chunk_size}")
```

```python
raw_chunk = ""
for chunk in chunks:
    print(chunk.shape)
    raw_chunk = chunk
    break;
```

```python
chunk = raw_chunk
```

```python
#Score = unkown => mean
chunk["Score"] = chunk["Score"].apply(lambda x:  np.NaN if x == "Unknown" else float(x))
chunk["Score"].fillna(chunk["Score"].mean(), inplace=True)

#Score-(1-10) = unkown => mean
for i in range(1,11):
    chunk[f"Score-{i}"] = chunk[f"Score-{i}"].apply(lambda x:  np.NaN if x == "Unknown" else float(x))
for i in range(1,11):
    chunk[f"Score-{i}"].fillna(chunk[f"Score-{i}"].mean(), inplace=True)
#Episodes = unkown => mean
chunk["Episodes"] = chunk["Episodes"].apply(lambda x:  np.NaN if x == "Unknown" else int(x))
chunk["Episodes"].fillna(chunk["Episodes"].mean(), inplace=True)

chunk.sort_values("Score" ,axis = 0,ascending = False).head(3)
```

```python
columns = ["Genres","Type","Source"]
for chunkColumn in columns:
    obj = {}
    obj["length"] = 0
    obj["columns"] = {}
    obj["columnsList"] = ['action', 'adventure', 'comedy', 'drama', 'sci-fi', 'space', 'mystery', 'shounen', 'police', 'supernatural', 'magic', 'fantasy', 'sports', 'josei', 'romance', 'slice of life', 'cars', 'seinen', 'horror', 'psychological', 'thriller', 'super power', 'martial arts', 'school', 'ecchi', 'vampire', 'military', 'historical', 'dementia', 'mecha', 'demons', 'samurai', 'game', 'shoujo', 'harem', 'music', 'shoujo ai', 'shounen ai', 'kids', 'hentai', 'parody', 'yuri', 'yaoi']

    def Genres(row, obj, isAdd):
        text = row[chunkColumn]
        # print(text)
        text = text.split(",")
        if isAdd:
            obj["length"] += len(text)
        if isAdd:
            for column in text:
                column = column.lower().strip()
                if column == "unknown": 
                    continue
                # print(obj["columns"][column])
                if column in obj["columns"].keys():
                    obj["columns"][column] +=  1
                else:
                    obj["columns"][column] = 1
        else:
            for column in text:    
                column = column.lower().strip()
                if column == "unknown": 
                    continue
                row[column] = round(obj["columns"][column] / obj["length"],3)
        return row

    chunk.apply(lambda x: Genres(x, obj, True),axis = 1)

    obj["columnsList"] = obj["columns"].keys()

    for column in obj["columnsList"]:
        if column == "unknown": 
            continue
        chunk[column] = np.zeros((chunk.shape[0], 1), dtype=int)


    chunk = chunk.apply(lambda x: Genres(x,obj,False),axis = 1)
```
